```python
from typing import Optional, Dict, Any
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings

def RAG_tool(query: str, filter: Optional[Dict[str, Any]] = None) -> str:
    vector_store = Chroma(
        embedding_function=OpenAIEmbeddings(model="text-embedding-3-large"), 
        persist_directory=r"C:\Users\jhwoo\Desktop\SKN_ws\project\SKN13-FINAL-1TEAM\chroma\kobaco_markdown"
    )

    retriever = vector_store.as_retriever(search_kwargs={"k":10})
    docs = retriever.invoke(query)

    result = "\n\n".join([doc.page_content for doc in docs])[:3000]
    return result
```

```python
query = input("query <<< ")
result = RAG_tool(query)
print(result)
```
